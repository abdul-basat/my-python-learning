## For loop

- 'a' or any other variable is used for loop to hold the number.
- 'range' is called generator, generates sequence of number
- generate numbers one after one not altogether
- will execute n-1 times.


```python
for a in range(10):
    print(a, 'Welcome to my web portal!')
```

```python
for num in range(5, 11):
    print(num, 'Pakistan Zindabad')
```

```python
print('Generating Odd number series:')
for num in range(1, 10, 2):    # (start, end-1, step)              #[1, 3, 5, 7, 9]
    print(num)
```

```python
print('Generating Even number series:')
for num in range(0, 10, 2):    # (start, end-1, step)              #[0, 2, 4, 6, 8]
    print(num)
```

```python
print('Generating Reverse series:')
for num in range(10, 0, -1):    # (start, end-1, step)
    print(num)
```

```python
print('Generating Reverse series: stepping 2')
for num in range(10, 0, -2):    # (start, end-1, step)              #[0, 2, 4, 6, 8]
    print(num)
```

```python
print('Generating Reverse series: stepping 2')
for num in range(9, 2, -2):    # (start, end-1, step)              #[0, 2, 4, 6, 8]
    print(num)
```

# Break and Continue

```python
# print -n in this case: i.e break just before the number

for a in range(10):
    if a == 5:
        break
    print(a)
```

```python
# when criteria completes, loop breaks:

for a in range(10):
    print(a)
    if a == 5:  
        break
```

```python
# print only the value and loop breaks:

for a in range(10):
    if a == 5:
        print(a)
        break
```

# continue
* we can skip some value on certain criteria by using <b><i>continue</i></b> keyword

```python
for a in range(10):
    if a == 5:
        continue
    print(a)
```

# While Loop

```python
a = 0 #counter variable
while a < 10: # loop iterate until it become True
    print(a)
    
# this code will generate infinite time as it has only 0 item. and every time it will result 0 and loops never breaks
```

```python
a = 0
while a < 10:
    a += 1
    print(a)
```

```python
a = 0
while a < 10:
    print(a)
    a += 1
```

```python
# reverse order:

a = 10
while a >= 0:
    print(a)
    a -= 1
```

```python
myguests = []
flag = True
while flag:
    guestName = input("Enter guest name or press q to exit: ")
    myguests.append(guestName)
    if guestName == "q":
        flag = False
myguests

# here q becomes part of the list because we appeneded it first and then breaks, we have to write condition first and then append
```

# List Comprehension

```python
evens = []
for num in range(20):
    if num%2 == 0:
        evens.append(num)
evens
```

```python
# list comprhension
[num for num in range(20) if num%2==0]
```

```python
odds = []
for num in range(20):
    if num%2 != 0:
        odds.append(num)
odds
```

```python
[num for num in range(20) if num%2!=0]
```

```python
# we can add more than one conditions in list comprehension e.g:
[num for num in range(20) if num%2!=0 if num>10]
```

### if else in list comprehension
- we can place if, else, elif and nested conditions in list comprehension
- only if will be placed after for loop but
- if else both should be placed first and then for loop

```python
[f"{num} is even" if num%2==0 else f"{num} is odd" for num in range(20)]
```

# Nested For Loop

```python
for table_num in range(2,3):
    for num in range(1, 11):
        print(f"{table_num} * {num} = {table_num * num}")
```

```python
[(x,y) for x in range(5) for y in 'ABCDE']
```

```python
[(x,y) for x in range(3) for y in 'ABC' if x>0]
```

```python
[a for a in range(1,50) if ((a%2==0) and (a%3==0))]
```

```python
a = []
for num in range(50):
    if num%2==0 and num%3==0:
        a.append(num)
print(a)
```

# Dictionary
- a dictioanry has a keyvalue pair
- dictioanry will retain the new value, if there are multiple keys of same name.
- {} is used for dictionary
- if {} is empty, its Dictionary, it has value, it is set.
- there is no indexing concept in dictionary


```python
student1 = {'name' : 'Jamal', 'Roll No' : '123', 'Email' : 'info@futeac.com'}
```

```python
student1
```

```python
student1['Roll No']
```

```python
student1['Email']
```

```python
# adding new members / new key in existing dictionary:
student1['father']= 'Hassan'
```

```python
student1
```

```python
# if we iterate directly on dictionary, it will returns keys only:
for a in student1:
    print(a)
```

```python
# if we iterate directly on dictionary.keys() , it will returns keys only:
for a in student1.keys():
    print(a)
```

```python
# We can get values by keys
for a in student1.keys():
    print(student1[a])
```

```python
# We can get values by keys
for a in student1.keys():
    print(f"{a} : {student1[a]}")
```

```python
# we can get values against kays by get()
for a in student1.keys():
    print(student1.get(a))
```

```python
# We can get values directly
for a in student1.values():
    print(a)
```

```python
# We can get values directly
for vals in student1.values():
    print(vals)
```
